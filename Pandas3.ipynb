### Apply func
#### It takes Prticular col and perform fun applied on each element of that col
#### Can work on particular row

```python
df.head(2)
```

```python
df["INR"]=df["Fare"].apply(lambda x : x *80)
df.head(2)
```

```python
df["Name_len"]=df["Name"].apply(len)
df.head(2)
```

```python
def test (x):
    if x<15:
        return " Teen"
    elif x>=15 and x<50:
        return "Young"
    else:
        return "Old"
        
```

```python
df["Age_grp"]=df["Age"].apply(test)
df.head(2)
```

```python
data = {"a":[1,2,3,4],
       "b":[4,5,6,7],
       "c":["sudh" , "krish","hitesh","navin"]}
df1 = pd.DataFrame(data,index = ['a','b','c','d'])
df1
```
