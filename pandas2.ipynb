### Different type of joins and set operstion in pandas

```python
df2=df[["Sex","Age"]][0:5]
df2
```

```python
df3=df[["Name","Sex","Age"]][8:12]
df3
```

```python
pd.concat([df2,df3])    # pd.concat() is just like set operation sql
                        # up-down set operation
```

```python
pd.concat([df2,df3],axis=1)   # same fun is used to perform set operation up-down and sidewise also
```

```python
df4=df[["Sex","Age"]][0:5]
df4
```

```python
df5=df[["Name","Sex","Age"]][3:8]
df5
```

```python
pd.concat([df4,df5])
```

```python
pd.concat([df4,df5],axis=1)
```

### join on a particular common col like sql join operation

##### pd.merge() is used

```python
df6=pd.DataFrame({ "id":[101,102,103],
                  "Name":["vis","ashis","prash"]
                 })
df6
```

```python
df7=pd.DataFrame({ "id":[101,104,105],
                  "Dpt":["mech","cse","Ai"]
                 })
df7
```

```python
pd.merge(df6,df7 , on="id")   # perform inner join based on commmon col eg id  (foreign key)
```

```python
pd.merge(df6,df7,on="id")   # same thing as above
```

```python
pd.merge(df6,df7,how="left")
```

```python
# pd.merge(df6,df7,on="Not_common_col_in_both_table") # will give error
```

```python
df8=pd.DataFrame({ "id":[101,102,103],
                  "Name":["vis","ashis","prash"]
                 },
                 index = ["a","b","c"])
df8
```

```python
df9=pd.DataFrame({ "deptid":["d10","d85","d51"],
                  "Dept":["mech","cse","Ai"]
                 },
                 index = ["a","e","f"])
df9
```

```python
df8.join(df9)  # perform left join autoimatically
               # same thing like sql join but on common rows
```

```python
df8.join(df9,how="inner")
```
